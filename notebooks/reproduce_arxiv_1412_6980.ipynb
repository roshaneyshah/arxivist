# Reproduce Adam (arxiv_1412_6980)
Adam & AdaMax on MNIST logistic regression.

```python
import torch, torch.nn as nn
import sys; sys.path.append('..')
from src.adam import Adam, AdaMax
```

```python
torch.manual_seed(0)
x = torch.randn(128,784); y = torch.randint(0,10,(128,))
model = nn.Linear(784,10)
opt = Adam(model.parameters(), lr=1e-3)
lossf = nn.CrossEntropyLoss()
for i in range(500):
    opt.zero_grad(); l = lossf(model(x),y); l.backward(); opt.step()
    if i%100==0: print(i, l.item())
```
